```python
import os

import pandas as pd

from kcmc_instance import KCMC_Instance

from tqdm.notebook import tqdm
```

```python
instances = pd.read_csv('instances.csv', sep='|', header=None)
instances.columns = ['serial', 'kcmc']
len(instances), instances.columns
```

```python
suffix = ' selected'
for _, row in tqdm(instances.iterrows(), total=len(instances)):
    kcmc = KCMC_Instance(row['serial'], True, True)
    filename = (
        kcmc.key_str.replace('_0.1:', ':').replace(':', '_')
      + f'_{kcmc.random_seed}.json'
    )
    if not os.path.exists('/data/visualizations/'+filename):
        rows = kcmc.cytoscape('/data/visualizations/'+filename)
        
    print(f'        <option value="{filename}"{suffix}>p{kcmc.num_pois}i{kcmc.num_sensors}s1 {kcmc.random_seed}</option>')
    suffix = ''
```
